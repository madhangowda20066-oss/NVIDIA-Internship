### Creating the resume PDF file

```python
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas

resume_content = """
DARSHINI

Email: [darshini@example.com](mailto:darshini@example.com)
Phone: +91 XXXXX XXXXX
Location: India

CAREER OBJECTIVE

Aspiring AI and Machine Learning Engineer seeking opportunities to apply skills in Python, Data Science, Machine Learning, and Deep Learning to solve real-world problems.

EDUCATION

Bachelor of Technology
Computer Science and Engineering
Expected Graduation: 2026

TECHNICAL SKILLS

Programming Languages:
Python, Java, SQL

Machine Learning:
Machine Learning, Deep Learning, Neural Networks

Frameworks & Tools:
TensorFlow, Scikit-learn, Pandas, NumPy, Matplotlib

Web Technologies:
HTML, CSS, JavaScript

Database:
MySQL

Cloud & Tools:
Git, GitHub, Google Colab

PROJECTS

1. AI Resume Screening & Job Recommendation System

* Developed an NLP-based application to analyze resumes and recommend suitable job roles.
* Implemented TF-IDF Vectorization and Cosine Similarity.
* Generated job recommendations with matching scores.

2. Student Performance Prediction

* Built a Machine Learning model to predict student academic performance.
* Performed data preprocessing, feature scaling, and model evaluation.

3. Face Recognition Attendance System

* Developed an automated attendance system using OpenCV and Face Recognition.
* Generated attendance reports and recognition results.

CERTIFICATIONS

* Python Programming
* Machine Learning Fundamentals
* Deep Learning Basics

SOFT SKILLS

* Problem Solving
* Communication
* Teamwork
* Leadership

LANGUAGES

* English
* Telugu
* Hindi
"""

def create_resume_pdf(filename, content):
    c = canvas.Canvas(filename, pagesize=letter)
    textobject = c.beginText()
    textobject.setTextOrigin(50, 750) # Start position
    textobject.setFont("Helvetica", 10)

    # Split content into lines and add to PDF
    lines = content.split('\n')
    for line in lines:
        textobject.textLine(line)

    c.drawText(textobject)
    c.save()

resume_filename = "Darshini.Resume.pdf"
create_resume_pdf(resume_filename, resume_content)

print(f"Resume saved as {resume_filename}")
```

```python
import pandas as pd
import pdfplumber
import matplotlib.pyplot as plt

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
job_data = {
    'Job_Role':[
        'Data Scientist',
        'Machine Learning Engineer',
        'AI Engineer',
        'Software Developer',
        'Data Analyst',
        'Frontend Developer',
        'Backend Developer',
        'Cloud Engineer'
    ],

    'Skills':[
        'Python Machine Learning SQL Statistics',
        'Python TensorFlow Deep Learning',
        'Python NLP Computer Vision TensorFlow',
        'Java OOP SQL',
        'Python Excel SQL PowerBI',
        'HTML CSS JavaScript React',
        'Java Spring Boot MySQL',
        'AWS Docker Kubernetes Python'
    ]
}

jobs = pd.DataFrame(job_data)

jobs.head()
```

```python
from google.colab import files

uploaded = files.upload()
```

```python
resume_file = list(uploaded.keys())[0]

resume_text = ""

with pdfplumber.open(resume_file) as pdf:
    for page in pdf.pages:

        txt = page.extract_text()

        if txt:
            resume_text += txt

print(resume_text[:2000])
```

```python
skills = [

'Python',
'Machine Learning',
'Deep Learning',
'TensorFlow',
'SQL',
'Java',
'HTML',
'CSS',
'JavaScript',
'React',
'Spring Boot',
'PowerBI',
'AWS',
'Docker',
'Kubernetes',
'NLP',
'Computer Vision'
]
```

```python
resume_skills = []

for skill in skills:

    if skill.lower() in resume_text.lower():

        resume_skills.append(skill)

print("Detected Skills:")
print(resume_skills)
```

```python
documents = jobs['Skills'].tolist()

vectorizer = TfidfVectorizer()

job_vectors = vectorizer.fit_transform(documents)

resume_vector = vectorizer.transform(
    [' '.join(resume_skills)]
)

similarity = cosine_similarity(
    resume_vector,
    job_vectors
)

jobs['Match_Score'] = similarity[0]

jobs.head()
```

```python
recommendations = jobs.sort_values(
    by='Match_Score',
    ascending=False
)

recommendations
```

```python
resume_score = recommendations['Match_Score'].max()*100

print(
    "Resume Score:",
    round(resume_score,2),
    "%"
)
```

```python
top_job = recommendations.iloc[0]

print("Recommended Job:")
print(top_job['Job_Role'])
```

```python
plt.figure(figsize=(10,5))

plt.bar(
    recommendations['Job_Role'],
    recommendations['Match_Score']
)

plt.xticks(rotation=45)

plt.title(
    "Job Recommendation Scores"
)

plt.xlabel(
    "Job Role"
)

plt.ylabel(
    "Similarity Score"
)

plt.show()
```

```python
top5 = recommendations.head(5)

plt.figure(figsize=(6,6))

plt.pie(
    top5['Match_Score'],
    labels=top5['Job_Role'],
    autopct='%1.1f%%'
)

plt.title(
    "Top Recommended Jobs"
)

plt.show()
```

```python
recommendations.to_csv(
    "Job_Recommendations.csv",
    index=False
)

print("Saved Successfully")
```
